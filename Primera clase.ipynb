```python
import requests
import pandas as pd
```

```python
url_ST = f"http://stranger-things-api.fly.dev/api/v1/characters?perPage=100&page=2"
peticion = requests.get(url_ST)
print(peticion.status_code)
```

```python
peticion.content
```

```python
peticion.text
```

```python
datos_ST = peticion.json()
for personaje in datos_ST:
    print(personaje.get('name','Unknown'))
```

```python
claves = []
for personaje in datos_ST:
    for clave in personaje.keys():
        if clave not in claves:
            claves.append(clave)
print(claves)
```

```python
datos_personaje_ST = []
for personaje in datos_ST:
    datos_personaje_ST.append({'name': personaje.get('name', 'Unknown'), 'status': personaje.get('status', 'Unknown'), 'residence': personaje.get('residence', 'Unknown')})
datos_personaje_ST
```

```python
dataframe_ST = pd.DataFrame(datos_personaje_ST)
dataframe_ST
```

```python
url_pokemon = f"https://pokeapi.co/api/v2/pokemon/"
llamada = requests.get(url_pokemon)

if llamada.status_code == 200:
    
    datos_pokemons = llamada.json()
    print("La llamada fue exitosa")
else:
    print("La llamada falló")

print(llamada.status_code)
```

```python
datos_pokemons
```

```python
url_2 = f"https://pokeapi.co/api/v2/pokemon/?offset=0&limit=100"
llamada_2 = requests.get(url_2)
datos_pokemons_2 = llamada_2.json()
datos_pokemons_2
```

```python
len(datos_pokemons_2["results"])
```

```python
name = []
for pokemon in datos_pokemons_2["results"]:
    name.append(pokemon["name"])
print(name)
print(len(name))
```

```python
dataframe_pokemons = pd.DataFrame(name)
dataframe_pokemons
```
