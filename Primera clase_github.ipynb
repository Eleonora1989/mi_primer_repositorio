```python
# cómo puedo buscar personajes mediante la URL
# Hacemos una petición al API
url = "https://swapi.py4e.com/api/people"
lista_personajes = []
# Se utiliza un bucle while para recorrer todas las páginas de la API usando el campo next, y un bucle for para almacenar cada personaje en una lista.
# Mientras haya una página siguiente
while url:
    response = requests.get(url)
    # Verificamos si la petición fue exitosa (código 200)
    if response.status_code == 200:
        # Mostramos los datos del API
        print(response.status_code)
        datos_api = response.json()
        print(datos_api)
        
        for personajes in datos_api["results"]:
            lista_personajes.append(personajes["name"]) 
            lista_personajes.append(personajes["url"]) 
  
        # Actualizamos la URL con la siguiente página
        url = datos_api["next"]

    else:
        # Mostramos un mensaje si hubo un error
        print(f"Error en la petición. Código de estado: {response.status_code}")

# Mostramos resultados
print(f"Total de personajes: {len(lista_personajes)}")
lista_personajes
```
